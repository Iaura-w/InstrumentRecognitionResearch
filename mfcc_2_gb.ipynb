```python
import os
import librosa
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data_mfcc_piano_violin.csv')
```

```python
df.dropna()
```

```python
from sklearn.preprocessing import StandardScaler

scaler =  StandardScaler()

for i in range(20):
    reshaped = np.array(df[str(i)]).reshape(-1, 1)
    df[str(i)] = scaler.fit_transform(reshaped)
```

```python
df
```

```python
from sklearn.model_selection import train_test_split

train_set, test_set = train_test_split(df, test_size=0.2, stratify=df['label'], random_state=42)
```

```python
test_set['label'].value_counts() / len(test_set)
```

```python
test_set.shape, train_set.shape
```

```python
X_train = train_set.drop(['label'], axis=1)
y_train = train_set['label'].copy()
```

```python
X_train.head()
```

```python
y_train.head()
```

```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()
labels_enc = label_encoder.fit(y_train)
```

```python
labels_enc
```

```python
label_encoder.classes_
```

```python
y_train = label_encoder.transform(y_train)
```

```python
y_train
```

```python
X_test = test_set.drop(['label'], axis=1)
y_test = test_set['label'].copy() 
y_test = label_encoder.transform(y_test)
```

```python
num_labels = np.unique(y_train).shape[0]
num_labels
```

```python
# models
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import make_scorer, accuracy_score 
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score
from sklearn.metrics import recall_score
from sklearn.metrics import classification_report
```

```python
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.model_selection import cross_val_predict
```

```python
# display_labels = ['piano','violin']
display_labels = label_encoder.inverse_transform(range(len(label_encoder.classes_)))
display_labels
```

```python
# gbc
```

```python
from sklearn.ensemble import GradientBoostingClassifier

model_gbc = GradientBoostingClassifier()

# parameters_gbc = {
#     'n_estimators': [100, 200, 250, 300, 400],
#     'learning_rate': [0.01, 0.1, 0.2, 0.5, 0.8, 1],
#     'max_depth': [3, 4, 5],
#     'min_samples_split': [2, 4, 6],
#     'min_samples_leaf': [1, 2, 4],
#     'max_features':[ 'log2', 'sqrt'],
# }

# grid_gbc = GridSearchCV(model_gbc, parameters_gbc, scoring=make_scorer(accuracy_score))
# grid_gbc.fit(X_train, y_train)

# model_gbc = grid_gbc.best_estimator_

model_gbc=GradientBoostingClassifier(learning_rate= 0.5,
 max_depth= 4,
 max_features= 'log2',
 min_samples_leaf= 4,
 min_samples_split= 4,
 n_estimators= 100)

model_gbc.fit(X_train, y_train)
```

```python
# grid_gbc.best_params_
```

```python
# train
```

```python
cv_score_gbc_train = cross_val_score(model_gbc, X_train, y_train, cv=5, scoring='accuracy')
cv_score_gbc_train
```

```python
print('Accuracy cross validation Gradient Boosting Train:', cv_score_gbc_train.mean())
```

```python
y_train_pred_gbc = cross_val_predict(model_gbc, X_train, y_train, cv=5)
ConfusionMatrixDisplay.from_predictions(y_train, y_train_pred_gbc, normalize='true', values_format='.0%', display_labels=display_labels)
plt.show()
```

```python
# test
```

```python
cv_score_gbc_test = cross_val_score(model_gbc, X_test, y_test, cv=5, scoring='accuracy')
cv_score_gbc_test
```

```python
print('Accuracy cross validation Gradient Boosting Test:', cv_score_gbc_test.mean())
```

```python
y_test_pred_gbc = cross_val_predict(model_gbc, X_test, y_test, cv=5)
ConfusionMatrixDisplay.from_predictions(y_test, y_test_pred_gbc, normalize='true', values_format='.0%', display_labels=display_labels)
plt.show()
```

```python
model_name = 'Gradient Boosting Test'
average = 'macro'

accuracy_score_gbc_test = accuracy_score(y_test, y_test_pred_gbc)
precision_score_gbc_test = precision_score(y_test, y_test_pred_gbc, average=average)
f1_score_gbc_test = f1_score(y_test, y_test_pred_gbc, average=average)
recall_score_gbc_test = recall_score(y_test, y_test_pred_gbc, average=average)

print(f'Accuracy {model_name}:', accuracy_score_gbc_test)
print(f'Precision {model_name}:', precision_score_gbc_test)
print(f'F1 score {model_name}:', f1_score_gbc_test)
print(f'Recall score {model_name}:', recall_score_gbc_test)
```

```python
re = classification_report(y_test, y_test_pred_gbc, target_names=display_labels, zero_division=0.0)

print(re)
```
